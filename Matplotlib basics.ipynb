```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as mticker
from matplotlib.finance import candlestick_ohlc

import numpy as np
import urllib
import matplotlib.dates as mdates
from matplotlib.dates import strpdate2num

MA1 = 5
MA2 = 20

#指標1
def moving_average(values, window):
    weights = np.repeat(1.0, window)/window
    smas = np.convolve(values, weights, 'valid')
    return smas

def high_minus_low(highs, lows):
    return highs-lows

#bytes轉成str形式判讀
def bytespdate2num(fmt, encoding='utf-8'):
    strconverter = mdates.strpdate2num(fmt) #注意，這是個bytes形式
    def bytesconverter(b):
        s = b.decode(encoding)
        return strconverter(s)
    return bytesconverter

#由於yahoo API被關閉，只能用一個當範例練習
def graph_data():
    
    #先定義figure，打上白底，不然儲存時可能顏色會變換
    fig = plt.figure(facecolor='#f0f0f0') 
    #創建圖像窗口，高10故意留間隔各一格
    ax1 = plt.subplot2grid((10,1),(0,0), rowspan=2, colspan=1) #plt.subplot2grid来创建第1个小图，(6,1)表示直6寬1，(0,0)表示從第0行第0列開始做圖
    plt.title('AAPL stock price')
    plt.ylabel('H-L')
    ax2 = plt.subplot2grid((10,1),(3,0), rowspan=4, colspan=1, sharex=ax1) #共享x軸
    plt.ylabel('Price')
    ax2v = ax2.twinx()
    
    ax3 = plt.subplot2grid((10,1),(8,0), rowspan=2, colspan=1, sharex=ax1) #共享x軸
    plt.ylabel('MAVGs')
    
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    source_code = urllib.request.urlopen(stock_price_url).read().decode()
    
    stock_data = []
    split_source = source_code.split("\n") #隔行的就分開成不同組資料
    
    #我想把資料變少一點
    count= 0
    while count<100:
        for line in split_source:
            split_line = line.split(',')
            if len(split_line) == 7:
                #幹他媽的，請把文字濾網加上去，不然格式會一直錯誤！！！
                if 'Date' not in line: 
                    stock_data.insert(0,line) 
                    #注意，sentdex後來給的資料跟以前的剛好相反，因此他用append加在後面，我要用insert放在前面才會跟他一樣
            count+=1
    #請再一次確認所獲得的是純數據！！
    #print(stock_data)
    
    Date, Open, High, Low, Close, Adjusted_close, Volume = np.loadtxt(stock_data,
                                                                      delimiter = ',', 
                                                                      unpack = True,
                                                                      converters = {0: bytespdate2num("%Y-%m-%d")})
    
    x = 1
    y = 21 #最終只想看計算20天
    ohlc = []
    
    while x < y:
        append_me = Date[-x], Open[-x], High[-x], Low[-x], Close[-x], Adjusted_close[-x], Volume[-x]
        ohlc.append(append_me)
        x+=1
    
    ma1 = moving_average(Close, MA1)
    ma2 = moving_average(Close, MA2)
    start = len(Date[0:MA2]-1)
    #print('start: '+str(start))
    
    h_l = list(map(high_minus_low, ma1[-start:], ma2[-start:]))
    
    #x,y跟下面一起標齊對正
    ax1.plot_date(Date[-start:], h_l[-start:], '-', label='H_L')
    ax1.xaxis.set_major_locator(mticker.MaxNLocator(nbins=5, prune='lower')) #bins是橫排數
    ax1.yaxis.set_major_locator(mticker.MaxNLocator(4)) #嘗試讓y軸不要擠在一起
    
    #幹你的，拜託看清楚自己data的樣子！！Sentdex新給的跟過去完全相反
    #print(len(ohlc))
    
    candlestick_ohlc(ax2, ohlc, width=0.4, colorup='red', colordown='g')
    ax2.xaxis.set_major_locator(mticker.MaxNLocator(nbins=5, prune='upper'))
    
    #此處是同時畫上成交量的圖
    ax2v.plot([], [], color='purple', alpha=0.4, label='Volume')
    ax2v.fill_between(Date[-start:], 0, Volume[-start:], facecolor='purple', alpha=0.5)
    ax2v.axes.yaxis.set_ticklabels([]) #不顯示y軸
    ax2v.grid(False) #不打格子線
    ax2v.set_ylim(0, 1.5*Volume.max()) #故意把y的上限拉的比最大值高很多，以縮小y看起來的規模
    
    
    #讓x,y可以對齊（因為有天數差異），不然不能做圖
    #print(len(Date), len(ma1))
    ax3.plot(Date[-start:], ma1[-start:], linewidth=1, label=(str(MA1)+'MA'))
    ax3.plot(Date[-start:], ma2[-start:], linewidth=1, label=(str(MA2)+'MA'))
    ax3.fill_between(Date[-start:], ma1[-start:], ma2[-start:], 
                     where=ma1[-start:] > ma2[-start:], facecolor='r', alpha=0.5)
    ax3.fill_between(Date[-start:], ma1[-start:], ma2[-start:], 
                     where=ma1[-start:] < ma2[-start:], facecolor='g', alpha=0.5)
    
    #把最重要的標線改到ax3
    ax3.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d')) #把unixtime轉換回日期格式
    ax3.xaxis.set_major_locator(mticker.MaxNLocator(10)) #把x的ticks放在好位置
    ax3.yaxis.set_major_locator(mticker.MaxNLocator(4)) #嘗試讓y軸不要擠在一起
    #finds up to a max number of ticks at nice locations
    ax3.xaxis.set_major_locator(mticker.MaxNLocator(nbins=5, prune='upper'))
    ax3.grid(True) #把格線確實打出來
    
    #把label傾斜角度
    for label in ax3.xaxis.get_ticklabels():
        label.set_rotation(45)
    #取消ax1, ax2的標線
    plt.setp(ax1.get_xticklabels(), visible=False)
    plt.setp(ax2.get_xticklabels(), visible=False)
    #最終修飾用，避免有label被吃掉的情形
    plt.subplots_adjust(left=0.12, bottom=0.2, right=0.9, top=0.90, wspace=0.20, hspace=0.20)
    

    #legend客製化，loc表示欄位位置，ncol表示要有幾個直排
    ax1.legend()
    leg = ax1.legend(loc=9, ncol=2, prop={'size':11})
    leg.get_frame().set_alpha(0.4)
    
    ax2v.legend()
    leg = ax2v.legend(loc=9, ncol=2, prop={'size':11})
    leg.get_frame().set_alpha(0.4)
    
    ax3.legend()
    leg = ax3.legend(loc=9, ncol=2, prop={'size':11})
    leg.get_frame().set_alpha(0.4)
    
    #前台顯示
    plt.show()
    
    #最上面plt.figure的地方有寫顏色，儲存圖片
    fig.savefig('AAPL.png', facecolor=fig.get_facecolor())
    
graph_data()
```

# Basemap intro

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

#projection是看世界的方式
#用經緯度放大小地區，resolution表示地圖精確度（f表示full）
m = Basemap(projection='mill',
            llcrnrlat = 25,
            llcrnrlon = -130,
            urcrnrlat = 50,
            urcrnrlon = -60,
            resolution = 'l')

#畫上海岸線
m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b') #畫州
m.drawcountries(color='darkred') #畫國家
#m.fillcontinents() #填滿大陸區塊
#m.etopo() #畫製地理圖
#m.bluemarble()



#凸顯特定點位
NYClat, NYClon = 40.7127, -74.0059
xpt, ypt = m(NYClon, NYClat)
xs = []
xs.append(xpt)
ys = []
ys.append(ypt)
m.plot(xpt, ypt, 'c*', markersize = 15)

LAlat, LAlon = 34.05, -118.25
xpt, ypt = m(LAlon, LAlat)
xs.append(xpt)
ys.append(ypt)
m.plot(xpt, ypt, 'g^', markersize=15)

m.plot(xs, ys, color='r', linewidth=3, label='Flight 98')
m.drawgreatcircle(NYClon, NYClat, LAlon, LAlat, color='c', linewidth=3, label='Arc')

plt.legend(loc=4)
plt.title('Basemap tutorial')
plt.show()
```
